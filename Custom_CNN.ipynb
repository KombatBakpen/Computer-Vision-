# Title: Custom CNN model with Cifar 100 dataset

```python
from keras.datasets import cifar100
import tensorflow as tf
import numpy as np
import keras
from keras.models import Sequential, Model
from tensorflow.keras.utils import plot_model
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
import matplotlib.pyplot as plt
import pandas as pd

data = cifar100.load_data()
(train_images, train_labels), (test_images, test_labels) = cifar100.load_data()

train_images, test_images = train_images/255., test_images/255.
image = train_images[1]
plt.imshow(image)
plt.show()


```

```python
FULLY_CONNECT_NUM = 128
NUM_CLASSES = 100
batch_size=32

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32,3),padding='same'))
#put input_shape=(28, 28,1) for digit
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu',padding='same'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu',padding='same'))
model.add(MaxPooling2D((2, 2)))

model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(NUM_CLASSES, activation='softmax'))
# model.add(layers.MaxPooling2D((2, 2)))

model.summary()
# Visualize model architecture as an image
plot_model(model, to_file='cifar100_model_architecture_detailed.png', show_shapes=True, show_layer_names=True, rankdir='TB')

# Display the generated model architecture image
img = plt.imread('cifar100_model_architecture_detailed.png')  # Read the image
plt.figure(figsize=(10, 20))  # Adjust the figure size as needed
plt.imshow(img)
plt.axis('off')  # Turn off axis labels
plt.show()

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_images, train_labels,
                    epochs=100,
                    validation_data=(test_images, test_labels),
                    batch_size = batch_size,
                    verbose = 1)



# Plot accuracy
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy over Epochs')
plt.legend()
plt.grid(True)

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(test_loss)
print(test_acc)

# metrics_df = pd.DataFrame(history.history)
# metrics_df[["loss","val_loss"]].plot();
# metrics_df[["accuracy","val_accuracy"]].plot();
# plt.show()



```

# My custom architecture  
To reduce the overfitting and improve the validation accuracy, different methods were applied, and the following gave the best results:

1. **Data Augmentation:** This method applies rotation, shifting, zoom, and shear to improve data variability. It has been found in the literature to reduce overfitting.

2. **Dropout Layers:** Dropout rates of 0.1, 0.2, and 0.3 were applied to each block of the CNN architecture, and 0.2% was applied to the dense layer. This has also been reported in the literature to prevent over-reliance on specific neurons, thus improving model robustness and reducing overfitting.

3. **Batch Normalization:** Batch normalization was used to normalize activations after each convolution. It has been reported in the literature to improve training efficiency and reduce overfitting.

4. **Regularization:** L2 regularization with a value of 0.001 was applied to the output layer. This was used to penalize large weights and help with better generalization of the model.

# OBSERVATION:
1. **Using these methods, the training loss was more consistent with the validation loss throughout the training process. After 100 epochs, the training accuracy was 84%, and the validation accuracy was 65%**.

2. **This model was further used to make predictions using random images downloaded from the internet. It correctly predicted 4 out 4 images.**

```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras.regularizers import L2

# Hyperparameters
FULLY_CONNECT_NUM = 128
NUM_CLASSES = 100
batch_size = 32
epochs = 100
fixed_learning_rate = 0.0003


output_folder = 'output/results'
os.makedirs(output_folder, exist_ok=True)

# Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.2,
    shear_range=0.2,
    fill_mode='nearest'
)
datagen.fit(train_images)

model = Sequential()

# Block 1
model.add(Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3), padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.1))

# Block 2
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

# Block 3
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.3))

# Global Average Pooling and Dense layers with L2 Regularization and Bias
model.add(GlobalAveragePooling2D())
model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(NUM_CLASSES, activation='softmax', kernel_regularizer=L2(0.001), bias_initializer='zeros'))

model.summary()

# Visualize model architecture as an image
from tensorflow.keras.utils import plot_model
plot_model(model, to_file=os.path.join(output_folder, 'cifar100_model_architecture_detailed.png'), show_shapes=True, show_layer_names=True, rankdir='TB')

img = plt.imread(os.path.join(output_folder, 'cifar100_model_architecture_detailed.png'))
plt.figure(figsize=(8, 40))
plt.imshow(img)
plt.axis('off')
plt.show()

# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=fixed_learning_rate)
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    datagen.flow(train_images, train_labels, batch_size=batch_size),
    epochs=epochs,
    validation_data=(test_images, test_labels),
    verbose=1
)

# Evaluate on test data
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

# Save the trained model
model.save(os.path.join(output_folder, 'cifar100_model.h5'))
print(f"Model saved as 'cifar100_model.h5' in {output_folder}")

# Plot training and validation accuracy/loss curves
plt.figure(figsize=(12, 4))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy over Epochs')
plt.legend()
plt.grid(True)

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig(os.path.join(output_folder, 'training_validation_accuracy_loss.png'))
plt.show()
```
